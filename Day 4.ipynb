```python
# Task You have 5 students in a class. For each student, we know:
# name (string), marks (0–100, int), hours_studied_per_week (int)
#Program should:, Loop through all students ,Determine performance:
# marks >= 70 and hours >= 10 → "Excellent"
# marks >= 50 and hours >= 5 → "Good" Else → "Needs Improvement"
# Keep track of:Total marks, Average marks, Print a clean report for each student and a class summary at the end:
```

```python
students =[
    {"name": "Chris", "marks":  80, "hours": 12},
    {"name": "Amy", "marks": 60, "hours": 8},
    {"name": "Samuel", "marks": 50, "hours": 5},
    {"name": "Sandra", "marks": 77, "hours": 12},
    {"name": "Philip", "marks": 55, "hours": 7}
]
```

```python
for student in students:
    print(student["name"])
    print(student["marks"])
```

```python
for student in students:
    if student["marks"] >= 70 and student["hours"] >= 10:
        performance = "Excellent"
    elif student["marks"] >= 50 and student["hours"] >= 5:
        performance = "Good"
    else:
        performance = "Needs Improvement"

print(f"{student['name']} Performance: {performance}")
```

```python
total_marks = 0
total_marks += student["marks"]
```

```python
average_marks = total_marks / len(students)
print(f"Class Average Marks: {average_marks}")
```

```python
print("----- Student Report -----")
print(f"Name: {student['name']}")
print(f"Marks: {student['marks']}")
print(f"Hours Studied: {student['hours']}")
print(f"Performance: {performance}")
print("--------------------------")
```

```python
print("----- Class Summary -----")
print(f"Total Marks: {total_marks}")
print(f"Average Marks: {average_marks:.2f}")
print("--------------------------")
```
