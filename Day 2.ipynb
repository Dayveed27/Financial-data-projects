```python
name = "Zion"
cgpa= 4.1
family_income = 45000

# Decision logic
if cgpa >= 3.7 and family_income < 30000:
    scholarship_status = "Full Scholarship"
elif cgpa >= 3.0 and family_income < 50000:
    scholarship_status =  "Partial Scholarship"
else:
    scholarship_status =  "Not eligible"
if scholarship_status == "Full Scholarship" :
    remarks = "Congratulations!"
elif scholarship_status == "Partial Scholarship" :
    remarks = "Keep Pushing"
else:
    remarks = "Don't give up"

# Clean Report
print("--------------------------------")
print(f"Name: {name}")
print(f"Cgpa: {cgpa}")
print(f"Family Income: {family_income}")
print(f"Scholarship Status: {scholarship_status}")
print(f"Remarks: {remarks}")
print("-------------------------------")
```
